```python
import pandas as pd
df = pd.read_csv("titanic.csv") 
df.head()
```

```python
print(df.isnull().sum())
```

```python
age_median = df['age'].median()
df['age'].fillna(age_median, inplace=True)

town_mode = df['embark_town'].mode()[0]
df['embark_town'].fillna(town_mode, inplace=True)

deck_mode = df['deck'].mode()[0]
df['deck'].fillna(deck_mode, inplace=True)

print(df.isnull().sum())
```

```python
age_median = df['age'].median()
df['age'] = df['age'].fillna(age_median)

town_mode = df['embark_town'].mode()[0]
df['embark_town'] = df['embark_town'].fillna(town_mode)

deck_mode = df['deck'].mode()[0]
df['deck'] = df['deck'].fillna(deck_mode)

print(df.isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['age', 'fare']] = scaler.fit_transform(df[['age', 'fare']])

df.head()
```

```python
cols_to_fix = ['sex', 'embark_town', 'who']

df = pd.get_dummies(df, columns=cols_to_fix, drop_first=True)

if 'alive' in df.columns:
    df.drop('alive', axis=1, inplace=True)

df.head()
```
